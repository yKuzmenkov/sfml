```python
# (8) построить среднее значение кредитного лимита для каждого вида образования 
# и для каждого пола
# график необходимо сделать очень широким (на весь экран)
_, ax_aa = plt.subplots(figsize=(20,8))

df.pivot_table(
    index = 'EDUCATION',
    columns = 'SEX',
    values = 'LIMIT_BAL',
    aggfunc = 'median'
).plot(kind = 'bar', stacked = True, ax = ax_aa)

plt.show()
```

```python
# (9) построить зависимость кредитного лимита и образования только для одного из полов

df['LIMIT_BAL'][df.SEX == 2].hist(alpha = 0.6, bins=20)
df['LIMIT_BAL'][df.SEX == 1].hist(alpha = 0.6, bins=20)
```
